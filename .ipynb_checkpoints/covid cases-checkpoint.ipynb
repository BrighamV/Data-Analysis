```python
import pandas as pd
from matplotlib import pyplot as plt
df = pd.read_csv('worldometer_coronavirus_daily_data.csv')
print(df.head())
```

```python
df = df.rename(columns={'daily_new_cases': 'cases', 'daily_new_deaths': 'deaths'})
print(df)
```

```python
df = df[['country', 'cases', 'deaths']]
print (df)
```

```python
df = df.groupby(['country']).sum()
print (df)
```

```python
df = df.sort_values('cases', ascending=False)
print(df)
```

```python
df = df.head(10)
print(df)
```

```python
countries = pd.read_csv('population_by_country_2020.csv')
print(countries.head())
```

```python
countries = countries.rename(columns={'Country (or dependency)': 'Country','Population (2020)': 'Population', 'Density (P/Km²)': 'Density'})
print(countries)
```

```python
countries = countries[['Country', 'Population', 'Density']]
print (countries)
```

```python
countries = countries.loc[(countries['Country'] == 'United States') | (countries['Country'] == 'India') | (countries['Country'] == 'Brazil') | (countries['Country'] == 'United Kingdom') | (countries['Country'] == 'France') | (countries['Country'] == 'Russia') | (countries['Country'] == 'Turkey') | (countries['Country'] == 'Germany') | (countries['Country'] == 'Spain') | (countries['Country'] == 'Italy')]
print(countries)
```

```python
countries = countries.sort_values('Density', ascending=False)
print(countries)
```

```python
plt.figure(figsize=(15, 5))
plt.plot(df.cases)
plt.title("Cases by Country")
plt.ylabel("Cases X 10,000")
plt.xlabel("Country")
plt.show()
```

```python
plt.figure(figsize=(15, 5))
plt.plot(countries.Country, countries.Density)
plt.title("Country Density")
plt.ylabel("Density in (P/Km²)")
plt.xlabel("Country")
plt.show()
```

```python
countries = countries.sort_values('Population', ascending=False)
plt.figure(figsize=(15, 5))
plt.plot(countries.Country, countries.Population)
plt.title("Country Population")
plt.ylabel("Population X 100,000,000")
plt.xlabel("Country")
plt.show()
```
